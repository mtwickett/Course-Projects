## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
df['CodeRevHrs'].isnull().value_counts()
```

```python
df['CodeRevHrs'].median()
```

```python
df['CodeRevHrs'].dtypes
```

```python
df['CodeRevHrs'].replace(np.nan, 4.0, inplace=True)
df['CodeRevHrs'].isnull().value_counts()
```

```python
df['WorkWeekHrs'].isnull().value_counts()
```

```python
df['WorkWeekHrs'].median()
```

```python
df['WorkWeekHrs'].replace(np.nan, 40.0, inplace=True)
df['WorkWeekHrs'].isnull().value_counts()
```

```python
df['Age'].isnull().value_counts()
```

```python
age_median = df['Age'].median()
age_median
```

```python
df['Age'].replace(np.nan, age_median, inplace=True)
df['Age'].isnull().value_counts()
```

```python
len(df)
```

```python
iqr_code_rev = df['CodeRevHrs'].quantile(0.75) - df['CodeRevHrs'].quantile(0.25)
print(iqr_code_rev)
```

```python
lower_bound_code_rev = df['CodeRevHrs'].quantile(0.25) - 1.5 * iqr_code_rev
lower_bound_code_rev
```

```python
upper_bound_code_rev = df['CodeRevHrs'].quantile(0.75) + 1.5 * iqr_code_rev
upper_bound_code_rev
```

```python
count = 0
for i in df['CodeRevHrs']:
    if i < lower_bound_code_rev or i > upper_bound_code_rev:
        count += 1
print(count)
```

```python
iqr_work_week = df['WorkWeekHrs'].quantile(0.75) - df['WorkWeekHrs'].quantile(0.25)
print(iqr_work_week)
```

```python
lower_bound_work_week = df['WorkWeekHrs'].quantile(0.25) - 1.5 * iqr_work_week
lower_bound_work_week
```

```python
upper_bound_work_week = df['WorkWeekHrs'].quantile(0.75) + 1.5 * iqr_work_week
upper_bound_work_week
```

```python
count = 0
for i in df['WorkWeekHrs']:
    if i < lower_bound_work_week or i > upper_bound_work_week:
        count += 1
print(count)
```

```python
count = 0
for i, j in zip(df['CodeRevHrs'], df['WorkWeekHrs']):
    if i < lower_bound_code_rev or i > upper_bound_code_rev or j < lower_bound_work_week or j > upper_bound_work_week:
        count += 1
print(count)
```

```python
df_removed_outliers = df[(df['CodeRevHrs'] > lower_bound_code_rev) & (df['CodeRevHrs'] < upper_bound_code_rev) & (df['WorkWeekHrs'] > lower_bound_work_week) & (df['WorkWeekHrs'] < upper_bound_work_week)]
```

```python
len(df_removed_outliers)
```

```python
df.head()
```

```python
norm_age = (df_removed_outliers['Age'] - df_removed_outliers['Age'].min()) / (df_removed_outliers['Age'].max() - df_removed_outliers['Age'].min())
df_removed_outliers.plot(kind='scatter',
        x='WorkWeekHrs',
        y='CodeRevHrs',
        figsize=(14, 8),
        alpha=0.5,
        color='green',
        s=norm_age * 1000 + 10
        )
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_name2 = 'DatabaseDesireNextYear'

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name2)

df2 = pd.read_sql_query(QUERY,conn)
print(df2.iat[0,0])
```

```python
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) AS Total
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY COUNT(DatabaseDesireNextYear) DESC
LIMIT 5
"""
df2 = pd.read_sql_query(QUERY,conn)
df2
```

```python
df2_databases = df2.groupby('DatabaseDesireNextYear', axis=0).sum()
df2_databases
```

```python
df2_databases['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%',
                            startangle=90,
                            )
plt.title('Top 5 Databases Respondents Wish To Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age BETWEEN 30 AND 35
"""
df3 = pd.read_sql_query(QUERY,conn)
df3.head()
```

```python
df3['CodeRevHrs'].replace(np.nan, 4.0, inplace=True)
df3['CodeRevHrs'].isnull().value_counts()
```

```python
df3['WorkWeekHrs'].replace(np.nan, 40.0, inplace=True)
df3['WorkWeekHrs'].isnull().value_counts()
```

```python
medians = df3.groupby('Age', axis=0).median()
medians.head()
```

```python
medians.plot(kind='bar', stacked='True')
plt.title('Median Work Week Hrs and Code rev Hrs for Ages 30 - 35')
plt.ylabel('Hrs')
```
